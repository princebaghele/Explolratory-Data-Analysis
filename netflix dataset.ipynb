### Step 2: Loading IMDb Ratings Data
Here, we are loading our first dataset from a file named `title.ratings.tsv`.
* A `.tsv` file is similar to a CSV, but the columns are separated by "tabs" instead of commas.
* We use `pd.read_csv()` to open the file and read it into a table called a 'DataFrame' (named `df_ratings`).
* `df_ratings.info()` shows us a quick summary of the data, such as the total number of rows and what type of data is in each column. This helps us ensure the data loaded correctly.

```python
# TSV stands for tab-separated value
#  it is similar to a CSV file and commonly used for tranfering data between spreadsheets and databases when your data is separated by commas and tabs.
# In this, we have text, layers, and records with one line.

# Load IMDb datasets (TSV)
df_ratings = pd.read_csv("title.ratings.tsv", sep="\t", low_memory=False)

print("\nIMDb Ratings Dataset:")
df_ratings.info()
```

### Step 3: Loading and Cleaning IMDb Basics Data
Now we are loading a second dataset, `title.basics.tsv`, which contains basic details about movies and shows.
* To save computer memory, we are only loading the specific columns we care about (like the title type, name, and release year).
* We are 'cleaning' the data by telling the computer to only keep rows where the type is a `'movie'`.
* We also check that the `startYear` only contains numbers, and then we convert it to an integer (a whole number) so we can do math on it later.

```python
df_basics = pd.read_csv(
    "title.basics.tsv", 
    sep="\t", 
    usecols = ['titleType', 'tconst', 'primaryTitle', 'originalTitle', 'startYear'],
    dtype = {'startYear': 'object'},
    low_memory=False
)

df_basics = df_basics[df_basics['titleType'] == 'movie']
df_basics = df_basics[df_basics['startYear'].str.isnumeric()]
df_basics['startYear'] = df_basics['startYear'].astype(int)

df_basics.head()
```

### Step 5: Combining the IMDb Datasets
In this step, we are combining (or 'merging') the two datasets we loaded in Step 2 and Step 3 into a single table.
* We use a function called `pd.merge()` to do an 'inner join'.
* This means the computer will look at both tables, find relationships using the unique ID column named `tconst`, and put matching rows side-by-side into a new DataFrame called `ratings_and_titles`.

```python
ratings_and_titles = pd.merge(df_ratings, df_basics, on='tconst', how='inner')
ratings_and_titles.head()
```

### Step 6: Loading Netflix Data
Now we turn our attention to the Netflix dataset.
* We load `netflix_titles.csv` into a DataFrame.
* We clean the data slightly by dropping (removing) any rows that have missing release years using `.dropna()`.
* Finally, we make sure the `release_year` column is treated as a number rather than text.

```python
# Load Netflix dataset (CSV format)
netflix_titles = pd.read_csv("netflix_titles.csv")

# Drop rows with missing release_year
netflix_titles = netflix_titles.dropna(subset=['release_year'])

# Convert release_year to integer type for analysis
netflix_titles['release_year'] = netflix_titles['release_year'].astype(int)

# Preview first 5 rows
netflix_titles.head()
```

### Step 7: Standardizing Movie Titles
Because we eventually want to match Netflix movies with IMDb movies, we need to make sure their names are formatted exactly the same.
* Sometimes "The Matrix" and "the matrix" might be seen as different texts by the computer.
* To fix this, we create new columns (`title_low` and `primaryTitle_low`) where every letter of the movie title is converted to lowercase using `.str.lower()`.
* This guarantees a fair comparison later on.

```python
# Create lowercase versions of titles in both datasets
# This helps avoid mismatches due to capitalization differences
netflix_titles['title_low'] = netflix_titles['title'].str.lower()
ratings_and_titles['primaryTitle_low'] = ratings_and_titles['primaryTitle'].str.lower()

# Preview the updated columns
netflix_titles[['title', 'title_low']].head()
ratings_and_titles[['primaryTitle', 'primaryTitle_low']].head()
```

### What to do next?
Now that we have successfully loaded, cleaned, and standardized both the Netflix dataset and the IMDb dataset, the goal is to cross-reference them.

**Next Steps to pursue in your project:**
1. **Merge the Datasets:** Combine the `netflix_titles` table with our `ratings_and_titles` table. You can match them using the lowercase title columns (`title_low` and `primaryTitle_low`). Watch out for movies with identical names; you may want to ensure they also share the same release year.
2. **Handle Mismatches & Missing Data:** After the merge, inspect the resulting table. Some Netflix movies might not have an IMDb match. Decide your strategy for these scenarios (e.g., discard them or keep them without a rating).
3. **Perform Exploratory Data Analysis (EDA):** 
   * What are the most common genres on Netflix with an IMDb rating above 8.0?
   * Find the top directors or cast with exceptionally high average ratings.
   * Are more highly-rated movies recent releases or older classics?
4. **Data Visualization:** Use the seaborn (`sb`) and matplotlib (`plt`) tools you imported in Cell 1 to draw clear visualizations. For example, a bar chart of the top 10 rated Netflix movies.

```python
grouped = ratings_and_titles.groupby('primaryTitle_low').apply(
    lambda x: pd.Series({
        'IMDB_numvotes':x['numVotes'].sum(),
        'IMDB_rating' : np.average(x['averageRating'],weights=x['numVotes'])
    })
).reset_index()
grouped['IMDB_rating'] = grouped['IMDB_rating'].round(1)

grouped.head()
```

```python
final = pd.merge(
    netflix_titles,
    grouped,
    left_on='title_low',
    right_on='primaryTitle_low',
    how='left'
)

final = final.drop(['title_low','primaryTitle_low'], axis=1)

final.head()
```

```python
final_clean = final[[
    'show_id','type','title','director','cast','country',
    'date_added','release_year','rating','duration','listed_in',
    'IMDB_rating','IMDB_numvotes'
]]

final_clean = final_clean.drop_duplicates()

final_clean['IMDB_rating'] = final_clean['IMDB_rating'].fillna(0)
final_clean['IMDB_numvotes'] = final_clean['IMDB_numvotes'].fillna(0)

final_clean['IMDB_numvotes'] = final_clean['IMDB_numvotes'].astype(int)

final_clean.head()
```

```python
final_clean.to_csv("netflix_powerbi.csv", index=False)
```
